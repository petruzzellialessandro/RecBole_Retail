# GRU4Rec

```python
from recbole.quick_start import run_recbole, load_data_and_model
from recbole.trainer import Trainer
```

```python
run_recbole(model="GRU4Rec", config_file_list=['config.yaml'],  dataset='RECEIPT_LINES')
```

```python
import os

model_file = [os.path.join('saved', file_name) for file_name in os.listdir('saved') if 'GRU4Rec' in file_name][0]
config, model, dataset, train_data, valid_data, test_data = load_data_and_model(model_file)
```

```python
trainer = Trainer(config, model)
best_valid_score, best_valid_result = trainer.fit(train_data, valid_data)
# Evaluate the model on the test set
result = trainer.evaluate(test_data)
# Access specific evaluation metrics if needed
print(result)
```
